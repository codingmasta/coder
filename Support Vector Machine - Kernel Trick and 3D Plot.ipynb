```python
# The goal of this assignment is to use the kernel trick to find
# the classification boundary line for linear support vector machine algorithm application.
```

```python
# Import packages

import numpy as np 
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a dataframe of random numbers

data = pd.DataFrame()
data["x"] = np.random.normal(0, .5, 1000)
data["y"] = np.random.normal(0, .5, 1000)


# Create your circle of indicators
data["d"] = data.x**2 + (data.y)**2 < .2

plt.figure(figsize = (5,5))
plt.scatter(data["x"], data["y"], c = data["d"], s = 20)

plt.show()
```

```python
# 3D plot 

from mpl_toolkits.mplot3d import Axes3D

# Make our new variable 
data["x2_y2"] = data.x**2 + data.y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.scatter(xs = data.x, ys = data.y, zs = data.x2_y2, zdir = "z", c = data.d)
```

```python
# Import package to add axis to 3d plot to determine boundary in 3D plot

from mpl_toolkits.mplot3d import Axes3D

# Make our new variable
data["x2_y2"] = data.x**2 + data.y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
xx, yy = np.meshgrid(range(-5, 5, 1), range(-5, 5, 1))
zz = np.empty((10, 10))
zz.fill(0.2)
ax.plot_surface(xx, yy, zz, color = "k", alpha = .5)
ax.scatter(xs = data.x, ys = data.y, zs = data.x2_y2, zdir = "z", c = data.d)
```

```python
""" 
Finalize the generation of the boundary by 
transforming the data into a higher-dimensional space, finding the optimal hyperplane, 
and then transforming it back.

"""

circle1 = plt.Circle((0, 0), np.sqrt(.2), color = "k", alpha = .2)

fig, ax = plt.subplots()

ax.scatter(data["x"], data["y"], c = data["d"], s = 20)

ax.add_artist(circle1)
```
